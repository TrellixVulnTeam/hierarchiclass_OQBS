This noteboook cretes a lookup dictionary that makes it possible to extract L1 and L2 labels for any L3 label.

```python
import pandas as pd
import pickle
```

### Building the dictionary from the data

```python
df = pd.read_csv('../data/classification_dataset.csv')
```

```python
lookup_dic = dict()
for l3_label in df.l3.unique():
    l1_label = df[df['l3'] == l3_label]['l1'].sample(1).to_string(index=False).strip()
    l2_label = df[df['l3'] == l3_label]['l2'].sample(1).to_string(index=False).strip()

    lookup_dic[l3_label] = (l1_label, l2_label, l3_label)
```

```python
lookup_dic
```

### Dumping the dictionary into a pickle file

```python
filename = '../misc/hierarchy_lookup_dict.pkl'
with open(filename,'wb') as outfile:
    pickle.dump(lookup_dic,outfile)
```

### Loading pickle

```python
with open(filename,'rb') as infile:
    new_dict = pickle.load(infile)

new_dict
```
